```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
from termcolor import colored
import matplotlib.pyplot as plt 
import plotly.graph_objects as go
import plotly.figure_factory as ff
from sklearn.model_selection import train_test_split
```

```python
df_sleep = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

df_sleep.head().style.set_properties(**{'background-color':'#007FFF', 'color': '#E2EEF3'})
```

```python
shape=colored(df_sleep.shape, 'magenta', None, attrs=['blink'])
print("A dimensão do dataset é de:", shape)
```

```python
df_sleep.info()
```

```python
df_sleep.describe().style.background_gradient(cmap='Blues')
```

```python
df_sleep.describe(include='O').style.set_properties(**{'background-color': '#007FFF','color': '#E2EEF3'})
```

```python
plt.style.use('seaborn-white')
sns.pairplot(data=df_sleep.drop('Person ID', axis=1), hue='Sleep Disorder', palette='mako')
plt.legend()
plt.show()
```

```python
grouped_data = df_sleep.groupby('Sleep Disorder')['Gender'].value_counts().unstack()
ax = grouped_data.plot(kind='bar', figsize=(15, 7), color=['#007FFF', 'blue'])
plt.title('Relação entre (sexo) e (Distúrbio do Sono)')
plt.xlabel('Distúrbio do Sono')
plt.ylabel('Contagem')
plt.legend(title='Gênero')
plt.show()
```
